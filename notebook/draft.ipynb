# Split train-test

The split should not allow for data of an individual to be in both train and test, therefore the id will be used to help split the data in 80% train and 20% test.

To be able to split individuals which each have 3 observations, let's simply split the id, stratifying according to the target.

```python
from sklearn.model_selection import train_test_split
```

```python
id_class = df.iloc[:, -1]
```

```python
id_class = id_class.iloc[range(0, len(id_class), 3)]
```

id_class was sampled here as all id are in triplicates but sorted, hence we can simply pick one id out of 3.

```python
print(id_class)
```

```python
train_id, test_id = train_test_split(range(0, 252), test_size=0.2, random_state=42, stratify=id_class)
```

Let's briefly check that picking the id from df returns the correct output.

```python
df.loc[train_id]
```

Let's now create the train and the test, without forgetting to shuffle them.

```python
train = df.loc[train_id].sample(frac=1)
test = df.loc[test_id].sample(frac=1)
```

```python
train
```

Now we can separate observations from target to create X_train, X_test, y_train, y_test.

```python
X_train = train.iloc[:, :-1]
y_train = train.iloc[:, -1]
X_test = test.iloc[:, :-1]
y_test = test.iloc[:, -1]
```

```python
X_train
```

# Normalization

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
```

```python
X_train
```

Gender is an interesting feature, it is binary and should not be normalized.

```python
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train.iloc[:, 1:])
```

```python
X_train_sc
```

```python
X_train = X_train.astype(float)
```

```python
X_train.dtypes
```

```python
X_train.iloc[:, 1:] = X_train_sc
```

```python
X_train
```

```python
X_test_sc = sc.transform(X_test.iloc[:, 1:])
X_test = X_test.astype(float)
X_test.iloc[:, 1:] = X_test_sc
X_test
```

# Features Selection

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
selector = SelectKBest(score_func=f_classif, k=100)
```

```python
X_train_slct = selector.fit_transform(X_train, y_train)
```

```python
X_train_slct.shape
```

```python
list(selector.get_feature_names_out())
```

```python
X_train['f1']
```

```python
X_test_slct = selector.transform(X_test)
```

```python
X_test_slct.shape
```

# Model Training : all features

## SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
svc_clf = SVC()
```

```python
svc_clf.fit(X_train, y_train)
```

```python
y_pred = svc_clf.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm_plot = ConfusionMatrixDisplay(cm)
cm_plot.plot()
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
searched_params = {'kernel': ['poly','sigmoid','linear','rbf'], 'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001]}
```

```python
svc_clf_opt = GridSearchCV(estimator=SVC(), param_grid=searched_params)
svc_clf_opt.fit(X_train, y_train)
```

```python
print(svc_clf_opt.best_params_)
```

```python
svc_clf2 = SVC(C=svc_clf_opt.best_params_['C'], kernel=svc_clf_opt.best_params_['kernel'], gamma=svc_clf_opt.best_params_['gamma'])
svc_clf2.fit(X_train, y_train)
y_pred2 = svc_clf2.predict(X_test)
cm2 = confusion_matrix(y_test, y_pred2)
cm_plot2 = ConfusionMatrixDisplay(cm2)
cm_plot2.plot()
```

```python
print(classification_report(y_test, y_pred2))
```
